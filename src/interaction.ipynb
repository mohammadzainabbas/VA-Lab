## Dataset: Seattle Weather Data

In this tutorial, we will create visualizations to explore weather data for Seattle from NOAA. We have the daily records of weather conditions from January 1st, 2012 to December 31st, 2015.

The dataset consists of columns for the temperature (in Celsius), precipitation (in millimeters), wind speed (in meters/second), and weather type (sun, fog, drizzle, rain, or snow).



```python
weather = data.seattle_weather()
weather.head()
```

## 1. Basic Interactions: Panning, Zooming, Tooltips

**Panning and Zooming:** You can use the ``interactive()`` shortcut to panning and zoom on the chart.

**Tooltip:** Add ``tooltip`` in the encoding of the chart and specify the variable you want to show when hovering over the visual element.

```python
alt.Chart(weather).mark_point(opacity=0.5).encode(
 x='temp_min:Q',
 y='temp_max:Q',
 color='weather:N',
 tooltip='date:T'
)
```

## 2. Brushing and Selections

There are three basic types of selections available:

* Interval Selection: `alt.selection_interval()`
* Single Selection: `alt.selection_single()`
* Multi Selection: `alt.selection_multi()`

### Brushing

To add selection behavior to a chart, we create the selection object and use the `add_selection` method:

```python
# Step 1: Create a brush interval selection


alt.Chart(weather).mark_point().encode(
 x='temp_min:Q',
 y='temp_max:Q',
 color='weather:N', # Step 3: Add condition to highlight the selection data
 tooltip='date:T'
)
# Step 2: Bind the brush on the chart
```

The brush will select points on the plot and stored it in the `interval` object.

Then, we can use the `condition()` function to change an element of the chart based on that input.:

The `alt.selection_interval()` function takes a number of additional arguments.

* `encodings` allows us to control whether the selection covers x, y, or both. For example, `alt.selection_interval(encodings=['x'])`.
* `empty` lets us control whether empty selections contain all values, or none of the values. For example, `alt.selection_interval(empty='none')`.

### Single selection

The `alt.selection_single()` function allows the user to click on single chart objects to select them, one at a time.

```python
# Step 1: Create a single selection


alt.Chart(weather).mark_point().encode(
 x='temp_min:Q',
 y='temp_max:Q',
 color='weather:N', # Step 3: Add condition to highlight the selection data
 tooltip='date:T'
)
# Step 2: Bind the selection on the chart
```

The single selection allows other behavior as well. For example, we can set `nearest=True` and `on='mouseover'` to update the highlight to the nearest point as we move the mouse.

We can select the weather condition on the legend to bind the selection on the scatterplot.

`single = alt.selection_single(fields=['weather'], bind='legend')`

### Multiple selection

The `alt.selection_multi()` function is quite similar to the `single` function, except it lets multiple points be selected at once, while holding the shift key.

```python
# Step 1: Create a multiple selection


alt.Chart(weather).mark_point().encode(
 x='temp_min:Q',
 y='temp_max:Q',
 color='weather:N', # Step 3: Add condition to highlight the selection data
 tooltip='date:T'
)
# Step 2: Bind the selection on the chart
```

## 3. Interaction with Compound Charts

We can create the compound chart by combinding two or more charts together. The Altair API provide three functions to create [multi-panel and layered charts](https://altair-viz.github.io/user_guide/compound_charts.html).

* Layering: `chart1 + chart2` OR `alt.layer(chart1, chart2)`
* Horizontal concatenation: `chart1 | chart2` OR `alt.hconcat(chart1, chart2)`
* Vertical concatenation: `chart1 & chart2` OR `alt.vconcat(chart1, chart2)`

```python
alt.Chart(weather).mark_point().encode(
 x='temp_min:Q',
 y='temp_max:Q',
 color='weather:N',
 tooltip='date:T'
)
```

```python
points = alt.Chart(weather).mark_point().encode(
 alt.X('monthdate(date):T', title='Date'),
 alt.Y('temp_max:Q',
 title='Maximum Daily Temperature (C)',
 scale=alt.Scale(domain=[-5, 40])
 ),
 color=color,
 size=alt.Size('precipitation:Q', scale=alt.Scale(range=[5, 200]))
).properties(
 width=800,
 height=300
)

bars = alt.Chart(weather).mark_bar().encode(
 x='count()',
 y='weather:N',
 color=color
).properties(width=800)

points & bars
```
